```python
import spacy
```

```python
# Load pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
#process text
doc = nlp(text)
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Install NLTK resources if not already done
nltk.download('punkt')
nltk.download('punkt_tab')

# Word-level tokenization
tokens = word_tokenize(text)
print("Word-Level Tokens:", tokens)
```

```python
# Character-level tokenization
tokens = list(text)
print("Character-Level Tokens:", tokens)
```

```python
from tokenizers import Tokenizer

# Load a pre-trained BPE tokenizer
tokenizer = Tokenizer.from_pretrained("gpt2")  # GPT-2 uses BPE

# Tokenize a sample text
output = tokenizer.encode(text)
print("BPE Tokens:", output.tokens)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer

# Initialize tokenizer and trainer
tokenizer = Tokenizer(BPE())
trainer = BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])

# Training corpus
corpus = text

# Train the tokenizer
tokenizer.train_from_iterator(corpus, trainer)

# Tokenize input
bpe_tokens = tokenizer.encode(text).tokens
print("BPE Tokens:", bpe_tokens)
```

```python
from transformers import BertTokenizerFast

# Load a pre-trained WordPiece tokenizer (BERT uses WordPiece)
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

# Tokenize a sample text
tokens = tokenizer.tokenize(text)
print("WordPiece Tokens:", tokens)
```

```python
from transformers import AutoTokenizer

# Load a pre-trained SentencePiece tokenizer (e.g., T5 or ALBERT uses SentencePiece)
tokenizer = AutoTokenizer.from_pretrained("t5-small")

# Tokenize a sample text
tokens = tokenizer.tokenize(text)
print("SentencePiece Tokens:", tokens)
```

##Compare the outputs of each technique and discuss the advantages and limitations of each:

1. Word-Level Tokenization (NLTK)
Observed Output: The text is split into human-readable words and punctuation marks. The advantages are the tokens are words, which is how humans process language. The limitation on the other hand, the model can't handle words it hasn't seen in training. Rare words, misspellings, or new proper nouns would all be mapped to a single "unknown" token,  causing the model to lose their meaning entirely. Also, huge Vocabulary and no morphological relation also serve as a limitation.

2. Character-Level Tokenization
Observed Output: A very long list of single characters, including spaces and newlines
. The advantages are since all words are built from characters, there are no "unknown" tokens. The model can represent any word, misspelling, or even URL. Also, the vocabulary size is extremely small. The limitations, however are that  individual tokens (like `'A'` or `'p'`) have almost no meaning on their own. The model must learn to build word-level concepts from scratch, which is computationally very expensive. Also, a single sentence becomes a very long sequence of tokens. This is a major problem for models like Transformers, as processing time often scales with the square of the sequence length.


3. Sub-word Tokenization (BPE, WordPiece, SentencePiece)

These methods are the modern standard and represent a compromise between word-level and character-level tokenization. They solve the OOV problem while keeping sequences manageable and retaining semantic meaning.
IT advantages is that it can represent any new word by splitting it into known sub-words.Also, the model can understand that "running" and "walking" are related because they both end in the `ing` token. The tokens are not always human-readable is however a liitation.Also, the way words are split depends entirely on the corpus the tokenizer was trained on.

### Comparison of Sub-word Techniques:

* **BPE (Pre-trained `gpt2`):**

Byte-Pair Encoding merges the most frequent pairs of tokens. You'll see tokens like `ĠApple` (the `Ġ` represents a space) and `Ġstartup`. Rare words like `Ifihan` and `Olusheye` will be split (e.g., `["Ifi", "han"]` and `["Ol", "us", "heye"]`).

* **WordPiece (Pre-trained `bert-base-uncased`):**

Very similar to BPE, but it merges tokens based on what most increases the *likelihood* of the training data. Its signature is using `##` to mark tokens that are not the start of a word.


* **SentencePiece (Pre-trained `t5-small`):**

This model treats the text as a raw stream, and the **space is included in the token**. This is considered a "cleaner" approach as it doesn't require pre-tokenization.
    * **Observed Output:** `[" Apple"]` (note the ` ` space prefix), `[" is"]`, `[" looking"]`, `[" at"]`, `[" buy", "ing"]`, `[" If", "i", "han"]`.
